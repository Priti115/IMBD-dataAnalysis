## Classify Movies Based on Rating [Excellent, Good and Average ]

```python
data.columns
```

```python
def rating(rating):
    if rating >= 7 :
        return "Excellent"
    elif rating >= 6 :
        return "Good"
    else:
        return "Avrage"
```

```python
data['ratng_cat'] = data['Rating'].apply(rating)
data.head(10)
```

## Find unique values from Genre

```python
data.columns
```

```python
data['Genre']
```

```python
myList = []
for value in data['Genre']:
    myList.append(value.split(','))
```

```python
myList
```

```python
empList = []
for item in myList:
    for item1 in item:
        empList.append(item1)    
```

```python
empList
```

```python
uni_list = []
for item in empList:
    if item not in uni_list:
        uni_list.append(item)
```

```python
uni_list
```
